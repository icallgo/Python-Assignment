# Measure of Spread

```python
variance= df.select_dtypes(
    include=[np.number]).var(ddof=1).to_frame()
variance.rename(columns={0: 'Variance'}, inplace=True)

std= df.select_dtypes(
    include=[np.number]).std(ddof=1).to_frame()
std.rename(columns={0: 'Standard Deviation'}, inplace=True)

range = df.describe().loc[['min', 'max']].T
range .rename(columns={'min': 'Minimum', 'max': 'Maximum'}, inplace=True)

Measure_of_Spread = pd.concat([variance, std, range], axis=1)

Measure_of_Spread.reset_index().rename(columns={'index': 'Attribute'})
```

- Variance dan standard deviation dapat dilihat di kolom diatas
- Range itu rentang dari nilai Minimum sampai Maximum
